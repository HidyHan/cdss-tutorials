#Basic Data Types

```r
myFun <- function(x) {
  return(sin(x) + .07*x^2 - cos(x))
}

myFun(10)
myFun(c(1,2,3,4,5))
```

```r
?seq
```

```r
vec <- seq(from=-10, to = 10, by = .1)
##most basic plot
plot(vec,myFun(vec))
lines(vec,myFun(vec))
```
